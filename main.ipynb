##### metrics

```python
from sklearn import metrics

def mean_reciprocal_rank(y_test, predicted):
    score = []
    for product in y_test:
        mrr = 1 / (list(predicted).index(product) + 1) if product in predicted else 0
        score.append(mrr)

    return np.mean(score)
```

```python
def evaluate_user_pred(user_id,y_true,y_pred):
    print('user:', user_id)
    print('y:', y_true), print('pred:',y_pred)

    print('true predicted:',len(set(y_true) & set(y_pred)))

    print(f'acccuracy: {round(metrics.accuracy_score(y_true,y_pred)*100)}%')
    print(f'mrr: {round(mean_reciprocal_rank(y_true,y_pred),2)}')
```

#### Content-based filtering

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(df_products[0])

cosine_sim = np.dot(tfidf_matrix, tfidf_matrix.T)
```

```python
def get_recommendations_for_product(product_index):
    scores = cosine_sim[product_index].toarray()[0]

    top_10 = np.argsort(scores)[::-1][1:11] # skipping the first-self element

    return df_products[0].iloc[top_10]
```

```python
prod_id = 1
recommendations = get_recommendations_for_product(prod_id)

print('Joke:', df_products[0][prod_id], '\n')
print('First recommended joke:', recommendations.values[0])
```

```python
def get_recommendations_for_user(user_id):
    train = df_users.iloc[user_id].to_frame(name="y")

    is_nan = np.isnan(train.values.reshape(-1))
    prods_watched_sim_scores = cosine_sim[~is_nan].toarray() # prods

    prods_ids = np.where(~is_nan)[0]
    prods_weights = train.iloc[prods_ids].values # user ratings

    # sum_sim_scores = np.mean(prods_weights * prods_watched_sim_scores, axis=0)
    sum_sim_scores = np.dot(prods_weights.T, prods_watched_sim_scores)

    products_ids = np.argsort( sum_sim_scores.reshape(-1) )[::-1]

    mask = np.isin(products_ids, prods_ids)
    recommended_prods_ids = products_ids[~mask][:10]

    return df_products[0].iloc[recommended_prods_ids]
```

```python
user_id = 1

prods = df_users.iloc[user_id].fillna(0).values
liked_jokes = np.argsort(prods.reshape(-1))[::-1]

recommendations = get_recommendations_for_user(user_id)

print('Joke:', df_products[0][liked_jokes[0]], '\n')
print('Joke:', df_products[0][liked_jokes[1]], '\n')

print('First recommended joke:', recommendations.values[0])
```

```python
# check of accuracy
top_k = 5
for i in range(15):
  train = df_users_train.iloc[i].to_frame(name="y")
  test = df_users_test.iloc[i].to_frame(name="y")
  # add all the test products but hide the y
  tmp = test.copy()
  tmp["y"] = np.nan
  train = pd.concat([train,tmp])

  prods_feat = cosine_sim.toarray()
  user_prods = train.fillna(0).values.T

  user_prods.shape, prods_feat.shape

  user_feat = np.dot(user_prods, prods_feat)
  weights = user_feat / user_feat.sum()

  pred = np.dot(weights, prods_feat.T)

  test['yhat'] = pred[0][-len(test):]

  test = test[~test["y"].isna()].reset_index(drop=True)

  y_true = test['y'].sort_values(ascending=False).index.values[:top_k]
  y_pred = test['yhat'].sort_values(ascending=False).index.values[:top_k]

  if len(y_pred) > 0:
    evaluate_user_pred(i,y_true,y_pred),print()
```

#### Coloborative filtering - memory based filtering (model based approach - neural networks)

scaling for further work with networks

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(.5,1.))
scaled_data = scaler.fit_transform(df_users)

df_users = pd.DataFrame(data=scaled_data, columns=df_users.columns)

df_users_train = df_users.loc[:, :int_split-1]
df_users_test = df_users.loc[:, int_split:]
```

model building

```python
from tensorflow import keras
import tensorflow as tf
```

```python
df_users_train = df_users_train.stack(dropna=True).reset_index().rename(
    columns={0:'y','level_0':'user','level_1':'product'})

df_users_test = df_users_test.stack(dropna=True).reset_index().rename(
    columns={0:'y','level_0':'user','level_1':'product'})
```

```python
embed_size = 50
user_size, prod_size = df_users.shape[0], df_products.shape[0]

# Users
x_users_input = keras.layers.Input(shape=[1,])

x_users_embed = keras.layers.Embedding(input_dim=user_size, output_dim=embed_size,
                                       name='users_embed')(x_users_input)

x_users = keras.layers.Reshape(target_shape=[embed_size])(x_users_embed)

# Products
x_prods_input = keras.layers.Input(shape=[1,])

x_prods_embed = keras.layers.Embedding(input_dim=prod_size,output_dim=embed_size,
                                        name='prods_embed')(x_prods_input)

x_prods = keras.layers.Reshape(target_shape=[embed_size])(x_prods_embed)

# Product
xx = keras.layers.Dot(normalize=True, axes=1)([x_users,x_prods])

# Output
output = keras.layers.Dense(1)(xx)

# Compiling
model = keras.models.Model(inputs=[x_users_input, x_prods_input],
                           outputs=[output])
model.compile(optimizer='adam', loss='mean_absolute_error',
              metrics=['mean_absolute_percentage_error'])
```

```python
df_users_train.head()
```

```python
model.fit([df_users_train['user'], df_users_train['product']], df_users_train['y'], epochs=5,
          batch_size=32, validation_split=.3, shuffle=True)
```

```python
df_users_test['yhat'] = model.predict([df_users_test['user'], df_users_test['product']])
df_users_test
```

```python
top_k = 5

for i in range(15):
  test = df_users_test.loc[df_users_test['user'] == i]

  y_true = test['y'].sort_values(ascending=False).index.values[:top_k]
  y_pred = test['yhat'].sort_values(ascending=False).index.values[:top_k]

  if len(y_pred) > 0:
    evaluate_user_pred(i,y_true,y_pred), print()
```
